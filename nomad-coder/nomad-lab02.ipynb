```python
# Let's organize all together

# import modules
import requests as reqs
from bs4 import BeautifulSoup
import pandas as pd
import time

target_site = 'https://weworkremotely.com'
uri = '/remote-full-time-jobs'
job_list = []

# scrape page
def scrap_page(url):

    response = reqs.get(f"{target_site}{uri}")
    
    soup = BeautifulSoup(response.text, 'html.parser')
    jobs = soup.find(
        'section',
        class_='jobs').find_all('li')[1:-1]
    
    # job data
    job_data = []
    
    for job in jobs:
        title = job.find('span', class_='title').text
        link = job.find_all('a')[-1]
    
        if not link:
            link = 'no link'
        else:
            link = target_site + link['href']
    
        try:
            firm, position, location = job.find_all('span', class_='company')
            location = location.text
            
        except ValueError:
            firm, position = job.find_all('span', class_='company')
            location = 'no data'
    
        firm = firm.text
        position = position.text
    
        job_data.append([firm, title, position, location, link])


    return job_data


# get pages
def get_pages(url):
    response = reqs.get(target_site + uri)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    pages = len(soup.find('div', class_='pagination').find_all('span', class_='page')) + 1

    return pages


# main
pages = get_pages(target_site + uri)

if pages:
    for i in range(1, pages):
        url = f"{target_site}{uri}?page={i}"
        job_list += scrap_page(url)
        time.sleep(1)


job_df = pd.DataFrame(
    job_list, 
    columns=[
        'company', 
        'title', 
        'position', 
        'location',
        'link',]
)

job_df
```
