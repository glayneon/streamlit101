```python
# all together
from bs4 import BeautifulSoup
import requests as reqs
import time
import sys

# variables
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36',
}

keywords = [
    'flutter',
    'python',
    'golang',
]

target_site = 'https://remoteok.com'
uri_prefix = '/remote-'
uri_postfix = '-jobs'

job_data = []

for page in keywords:
    url = f"{target_site}{uri_prefix}{page}{uri_postfix}"
    response = reqs.get(url, headers = headers)

    if not response.status_code == 200:
        print(f"Error - {response.status_code}")
        sys.exit()

    soup = BeautifulSoup(response.text, 'html.parser')

    # remove ads with [1:]
    jobs = soup.find('table', id='jobsboard').find_all('td', class_='company position company_and_position')[1:]

    for job in jobs:
        link = job.find('a')['href']
        title = job.find('h2').text.strip('\n')
        company = job.find('h3').text.strip()
        location = job.find('div', class_='location').text.strip()

        try:
            salary = job.find('div', class_='location tooltip').text.strip()
        except AttributeError:
            salary = 'no data'

        job_data.append(
            [title, company, location, salary, link]
        )

    time.sleep(1)


for i in job_data:
    print(i)
```

```python
# class implementation
from bs4 import BeautifulSoup
import requests as reqs
import sys

# variables
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36',
}

example_keywords = [
    'flutter',
    'python',
    'golang',
]


class RemoteOK:

    def __init__(self, header):
        self.job_data = []
        # Specifying HERE-PATTERN on URL
        self.remoteok = 'https://remoteok.com/remote-HERE-PATTERN-jobs'
        self.header = header

    def scrap_page_wt(self, keyword):
        if type(keyword) is not str:
            print(f"Keyword {keyword} is not a type of STR.")
            sys.exit(0)

        url = self.remoteok.replace('HERE-PATTERN', keyword)

        print(f"Searching jobs - {keyword} on remoteok.com ...")
        response = reqs.get(url, headers = self.header)

        if not response.status_code == 200:
            print(f"Error - {response.status_code}")
            sys.exit(0)

        soup = BeautifulSoup(response.text, 'html.parser')

        # remove ads with [1:]
        jobs = soup.find('table', id='jobsboard'
                        ).find_all('td', class_='company position company_and_position')[1:]

        for job in jobs:
            link = job.find('a')['href']
            title = job.find('h2').text.strip('\n')
            company = job.find('h3').text.strip()
            location = job.find('div', class_='location').text.strip()

            try:
                salary = job.find('div', class_='location tooltip').text.strip()
            except AttributeError:
                salary = 'no data'

            self.job_data.append(
                [title, company, location, salary, link]
            )

    def show_jobs(self):
        if self.job_data:
            for job in self.job_data:
                print(job)
        else:
            print("No jobs found.")

    def save_jobs2csv(self):
        # next time...
        pass



jobsearching = RemoteOK(header = headers)

for keyword in example_keywords:
    jobsearching.scrap_page_wt(keyword)

jobsearching.show_jobs()

```
