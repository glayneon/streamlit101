```python
# from bs4 import BeautifulSoup
# 명령형 기능 및 결과 저장
# pypi 로 저장
import pandas as pd
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
# from subprocess import CREATE_NO_WINDOW
import time

title = []
price = []
keyword = '어센션'
pg = 1
max_pg = 5

driver = Chrome()

while pg < max_pg:
    page_url = f"https://boardlife.co.kr/shop.php?search={keyword}&pg={pg}"
    driver.get(page_url)
    time.sleep(2)

    pg += 1
    
    markets = driver.find_element(By.XPATH, '//*[@id="shop-wrapper"]')

    try:
        if markets.find_element(By.CLASS_NAME, 'no-shop-info'):
            break
            
    except:
        goods_list = markets.find_elements(By.CLASS_NAME, 'market-row')
        
        for good in goods_list:
            title.append(good.find_element(By.CLASS_NAME, 'title').text.split('\n')[1])
            price.append(good.find_element(By.CLASS_NAME, 'baro-price').text)

# Genarate DataFrame
bora = {
    'title': title,
    'price': price
}

bora_df = pd.DataFrame(bora)
driver.close()
bora_df
```
